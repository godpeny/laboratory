```python
# import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time

from google.colab import drive
```

```python
# using Google Drive
drive.mount('/d')
model_path = '/d/My Drive/Colab Notebooks/DL-tutorial/models/house/best-model.hdf5'
test_data_path = '/d/My Drive/Colab Notebooks/DL-tutorial/data/kaggle/house-prices-advanced-regression-techniques/test.csv'

# load test sets
test = pd.read_csv(test_data_path)

# load model made before
model = load_model(model_path)
```

```python
"""
preprocessing the test case
"""
test = pd.get_dummies(test) # one-hot encodding
test = test.fillna(test.mean()) # fill null value with mean of the each column

# choose columns to use
cols = ['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF'] # same as house model
Test = test[cols]
Test.head(5)
```

```python
"""
prediction with test sets
"""
result = []
prediction = model.predict((Test)).flatten()
print(prediction)

for i in range(len(prediction)):
  id = test['Id'][i]
  predict = prediction[i]
  result.append([id, predict])
```

```python
# save test results
output_dir = '/d/My Drive/Colab Notebooks/DL-tutorial/tests/house/'
timestr = time.strftime("%Y%m%d-%H%M%S")
file_name = str(timestr)

df = pd.DataFrame(result, columns=["Id", "SalePrice"])
df.to_csv(output_dir+file_name+"_submission.csv", index=False)
```
