```python
# import
import numpy as np
import random as rand
import matplotlib.pyplot as plt
```

```python
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])

w = rand.randint(-10, 10) # weight 기울기
b = rand.randint(-100, 100) # bias 절편

n = len(x)
lr = 0.03 # learning rate
epochs = 1001 # 반복 횟수
```

스크린샷 2023-09-09 21.11.49.png

```python
for i in range(epochs):
  py = (w * x) + b # predicted y
  err = py - y # error = 오차

  w_grad = (2/n) * sum(x * err) # a로 편미분 한 값 (편미분 공식 참조) = mean((py-y)*x) / (2/n) * sum == mean
  b_grad = (2/n) * sum(err) # b로 편미분 한 값

  w = w - (w_grad * lr) # 경사하강 : 미분값을 빼서 최소값 구함, 경사상승 : 미분값 더해서 최대값 구함
  b = b - (b_grad * lr)

  if i%100 == 0:
    print("[" + str(i) + "]" + " w : " + str(w) + " / b : " + str(b))
```

```python
py = (w*x) + b
plt.scatter(x,y)
plt.plot(x, py, 'r')
```
