```python
# import
import numpy as np
import random as rand
import matplotlib.pyplot as plt
```

```python
x1 = np.array([2,4,6,8])
x2 = np.array([0,4,2,3])
y = np.array([81,93,91,97])

lr = 0.01 # learning rate
epochs = 2001

w1 = 0 # weigh
w2 = 0 # weigh
b = 0 # bias
n = len(x1)

# Create a figure
fig = plt.figure()

# Add 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Plot scatter
ax.scatter(x1, x2, y)

# Add labels and title
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
ax.set_title('3D Scatter Plot')

# Show plot
plt.show()
```

```python
for i in range(epochs):
  py = (w1 * x1) + (w2 * x2) + b
  err = py - y

  w1_grad = np.mean(x1 * err) * 2 # (*2) could be ignored
  w2_grad = np.mean(x2 * err) * 2
  b_grad = np.mean(err) * 2

  w1 = w1 - (lr * w1_grad)
  w2 = w2 - (lr * w2_grad)
  b = b - (lr * b_grad)

  if i%100 == 0:
    print("[" + str(i) + "]" + " w1 : " + str(w1) + " w2 : " + str(2) + " / b : " + str(b))
```
