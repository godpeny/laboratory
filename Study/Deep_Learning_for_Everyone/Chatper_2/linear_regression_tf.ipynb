```python
# import
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential # Sequential groups a linear stack of layers into a tf.keras.Model
from tensorflow.keras.layers import Dense # Just your regular densely-connected NN layer.
```

```python
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])

model = Sequential()

model.add(Dense(1, input_dim=1, activation='linear')) # input_dim = # of input, activation function = linear regression
model.compile(optimizer='sgd', loss='mse') # Configures the model for training, sgd=Stochastic Gradient Descent, mse=Mean Square Error
model.fit(x, y, epochs=2000) # Trains the model for a fixed number of epochs
```

```python
plt.scatter(x,y)
plt.plot(x, model.predict(x), 'g') # model.predict : Generates output predictions for the input samples.
plt.show()
```

```python
predict = model.predict([7]) # predict when x is 7
print(predict)
```
